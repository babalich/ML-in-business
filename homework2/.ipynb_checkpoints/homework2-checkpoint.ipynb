```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import lightgbm as lgbm
```

```python
def exponential_smoothing(series, alpha):
    result = [series[0]] # first value is same as series
    for n in range(1, len(series)):
        result.append(alpha * series[n] + (1 - alpha) * result[n-1])
    return result
```

```python
def split_data_b( data, split_date ):
    return data.loc[data.index.get_level_values('Date') <= split_date].copy(), \
           data.loc[data.index.get_level_values('Date') >  split_date].copy()
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
data = pd.read_csv('my_PJME_MW.csv', index_col=[0,1], parse_dates=[0])
#data.head(3)
```

```python
data['exp_001'] = exponential_smoothing(data['PJME_MW'], 0.01)
data['exp_005'] = exponential_smoothing(data['PJME_MW'], 0.05)
data['exp_010'] = exponential_smoothing(data['PJME_MW'], 0.10)
#day_data.head(3)
```

```python
_date = '2016-01-01'
plt.figure(figsize=(16, 6))
plt.plot(data['PJME_MW'].loc[data.index.get_level_values('Date') >= _date].values, '.b', label='Исходный график', alpha=0.05)
#plt.plot(data['exp_001'].loc[data.index.get_level_values('Date') >= _date].values, color='g', label='Exp 0.01')
plt.plot(data['exp_005'].loc[data.index.get_level_values('Date') >= _date].values, color='g', label='Exp 0.05')
#plt.plot(data['exp_010'].loc[data.index.get_level_values('Date') >= _date].values, color='g', label='Exp 0.10')
plt.grid()
plt.legend()
```

```python
data['t1'] = 0
data['t1'].iloc[1:] = data['PJME_MW'].iloc[:-1].values

data['t2'] = 0
data['t2'].iloc[2:] = data['PJME_MW'].iloc[:-2].values

data['t3'] = 0
data['t3'].iloc[3:] = data['PJME_MW'].iloc[:-3].values

# for i in range(1, 50):
#     data[f't{i}'] = 0
#     data[f't{i}'].iloc[i:] = data['PJME_MW'].iloc[:-i].values
```

```python
data['diff1'] = data['t1'] - data['PJME_MW']
data['diff1'][0] = 0
```

```python
data.head()
```

```python
model = lgbm.LGBMRegressor(n_estimators=200, max_depth=15, random_state=10)
```

```python
split_point = '2016-01-01'
train, test = split_data_b(data, split_point)
```

```python
h = 1
X_train = train.iloc[:-h, :]
X_test = test.iloc[:-h,:]
y_train = train['exp_005'].values[h:]
y_test = test['exp_005'].values[h:]

model.fit(X_train, y_train)
X_test_pred_1 = model.predict(X_test)
```

```python
mean_absolute_percentage_error(y_test, X_test_pred_1)
```

```python
h = 10
X_train = train.iloc[:-h, :]
X_test = test.iloc[:-h,:]
y_train = train['exp_005'].values[h:]
y_test = test['exp_005'].values[h:]

model.fit(X_train, y_train)
X_test_pred_2 = model.predict(X_test)
```

```python
mean_absolute_percentage_error(y_test, X_test_pred_2)
```

```python
plt.figure(figsize=(16, 6))
plt.plot(data['PJME_MW'][-300:].values, '.b', label='Исходный график', alpha=0.4)
plt.plot(y_test[-300:], 'xr', label='test data')
plt.plot(X_test_pred_1[-300:], color='g', label='Predict h=1')
plt.plot(X_test_pred_2[-300:], color='orange', label='Predict h=10')
plt.grid()
plt.legend()
```

Мы видим значительный рост ошибки с увеличением глубины прогноза. Предсказания получаются менее сглаженные, визуально видно более острык пики (в основном вверх)
